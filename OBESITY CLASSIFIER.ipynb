```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay, classification_report
)
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE

def evaluate_all_models(models, param_grids, df, output_file, scalers, test_splits):
    results = []
    smote = SMOTE(random_state=42)  # For oversampling

    for scaler_name, scaler in scalers.items():
        for test_split in test_splits:
            print(f"Running with {scaler_name} scaler and test split {test_split}...")
            X = df.drop(columns=['new_labels'])
            y = df['new_labels']

            # Balance the dataset using SMOTE
            X_balanced, y_balanced = smote.fit_resample(X, y)

            # Split into training and testing datasets
            X_train, X_test, y_train, y_test = train_test_split(
                X_balanced, y_balanced, test_size=test_split, random_state=42, stratify=y_balanced
            )

            # Apply scaling
            X_train_scaled = scaler.fit_transform(X_train)
            X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
            X_test_scaled = scaler.transform(X_test)
            X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

            for model_name, model in models.items():
                print(f"Evaluating {model_name}...")

                param_grid = param_grids.get(model_name, {})
                param_combinations = list(itertools.product(*param_grid.values())) if param_grid else [None]

                for params in param_combinations:
                    model_instance = model.__class__()

                    if params:
                        param_dict = dict(zip(param_grid.keys(), params))
                        model_instance.set_params(**param_dict)
                    else:
                        param_dict = {}

                    model_instance.fit(X_train_scaled, y_train)

                    # Training predictions
                    y_pred_train = model_instance.predict(X_train_scaled)
                    y_pred_proba_train = (
                        model_instance.predict_proba(X_train_scaled)
                        if hasattr(model_instance, "predict_proba") and len(np.unique(y_train)) ==
                        model_instance.predict_proba(X_train_scaled).shape[1]
                        else None
                    )

                    # Testing predictions
                    y_pred_test = model_instance.predict(X_test_scaled)
                    y_pred_proba_test = (
                        model_instance.predict_proba(X_test_scaled)
                        if hasattr(model_instance, "predict_proba") and len(np.unique(y_test)) ==
                        model_instance.predict_proba(X_test_scaled).shape[1]
                        else None
                    )

                    # Calculate metrics
                    train_metrics = {
                        'Accuracy': accuracy_score(y_train, y_pred_train),
                        'Precision': precision_score(y_train, y_pred_train, average='weighted'),
                        'Recall': recall_score(y_train, y_pred_train, average='weighted'),
                        'F1': f1_score(y_train, y_pred_train, average='weighted'),
                        'AUC': roc_auc_score(y_train, y_pred_proba_train, multi_class='ovr') if y_pred_proba_train is not None else "N/A"
                    }
                    test_metrics = {
                        'Accuracy': accuracy_score(y_test, y_pred_test),
                        'Precision': precision_score(y_test, y_pred_test, average='weighted'),
                        'Recall': recall_score(y_test, y_pred_test, average='weighted'),
                        'F1': f1_score(y_test, y_pred_test, average='weighted'),
                        'AUC': roc_auc_score(y_test, y_pred_proba_test, multi_class='ovr') if y_pred_proba_test is not None else "N/A"
                    }

                    # Append results
                    results.append({
                        'Scaler': scaler_name,
                        'Test Split': test_split,
                        'Model': model_name,
                        'Parameters': param_dict,
                        **{f"Train {key}": value for key, value in train_metrics.items()},
                        **{f"Test {key}": value for key, value in test_metrics.items()}
                    })

                    # Generate Confusion Matrix Plots
                    # cm_train = confusion_matrix(y_train, y_pred_train)
                    # cm_test = confusion_matrix(y_test, y_pred_test)

                    # fig, ax = plt.subplots(1, 2, figsize=(12, 6))
                    # sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', ax=ax[0])
                    # ax[0].set_title(f"{model_name} - Training Confusion Matrix")
                    # sns.heatmap(cm_test, annot=True, fmt='d', cmap='Oranges', ax=ax[1])
                    # ax[1].set_title(f"{model_name} - Testing Confusion Matrix")
                    # # plt.savefig(f"confusion_matrix_{model_name}_{scaler_name}_{test_split}.png")
                    # plt.close()

    # Save results
    results_df = pd.DataFrame(results)
    results_df.to_excel(output_file, index=False)
    print(f"Results saved to {output_file}")
    return results_df



scalers = {
    'MinMax': MinMaxScaler(),
    'Standard': StandardScaler(),
    'MaxAbs': MaxAbsScaler()
}

test_splits = [0.2,0.3]

for i in range(1,5):
  labels=pd.read_csv(f'/content/drive/MyDrive/OBESITY/new_labels_Obesity_set{i}.csv')
  df_train_labeld=pd.merge(df_train,labels[['Patient_ID','new_labels']],on='Patient_ID')
  df_train_labeld=df_train_labeld.drop(columns=['Patient_ID'])
  df_results=evaluate_all_models(models, param_grids, df_train_labeld, f'/content/drive/MyDrive/OBESITY/results_Obesity_set{i}.xlsx', scalers, test_splits)
```
